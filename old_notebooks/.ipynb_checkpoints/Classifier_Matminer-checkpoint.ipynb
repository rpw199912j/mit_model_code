```python
def mean_imputation(X):
    imputed_df = X.fillna(X.mean())
    return imputed_df
```

```python
def read_data(choice, path):
    """Pick one class to predict. The other two will be lumped together into one class.
    
    Args:
        choice: String, 'MIT', 'Insulator', or 'Metal'
    Returns:
        X: DataFrame, features used in model
        y: Series, class labels
    """
    df = pd.read_excel(path)
    df_known = df.loc[df.Label.notnull()].copy()
 
    df_known.dropna(thresh=df_known.shape[1]-6, inplace=True) # Drop rows with at least six columns missing
    # Impute some missing data using mean
#     df_known = mean_imputation(df_known)
#     df_known.fillna(value={'esthubu': df_known['esthubu'].mean(), 'estct': df_known['estct'].mean(),
#                               'minimum oxidation state': df_known['minimum oxidation state'].mean(), 
#                                'maximum oxidation state': df_known['maximum oxidation state'].mean(),
#                                'range oxidation state': df_known['range oxidation state'].mean(),
#                                'std_dev oxidation state': df_known['std_dev oxidation state'].mean(),
#                                'ewald_energy': df_known['ewald_energy'].mean(),
#                            'v_mad_m': df_known['v_mad_m'].mean(),
#                            'v_mad_x': df_known['v_mad_x'].mean(),
#                            'ie1': df_known['ie1'].mean(),
#                            'ie2': df_known['ie2'].mean()
#                               }, inplace=True)
    # Use ML to impute missing GII values
#     gbr = GradientBoostingRegressor(n_estimators=20)
#     gbr.fit(df_known[~df_known['gii'].isnull()].drop(
#         columns=['Compound', 'gii', 'struct_file_path']), y=df_known[~df_known['gii'].isnull()]['gii'])
#     X_test = df_known.loc[df_known['gii'].isnull()].drop(columns=['Compound', 'gii', 'struct_file_path'])
#     df_known.loc[df_known['gii'].isnull(), 'gii'] = gbr.predict(X_test)
    
    s = df_known["Label"]
    if choice == "Insulator":
        cond = s == 1
        df_known["class"] = s.where(cond, 0)

    elif choice == "MIT":
        cond = s == 2
        df_known["class"] = s.where(cond, 0)
        df_known["class"].replace(2, 1, inplace=True)
    elif choice == 'Metal':
        cond = s == 0
        df_known["class"] = s.where(cond, 2)
        df_known["class"].replace(0, 1, inplace=True)
        df_known["class"].replace(2, 0, inplace=True)
    elif choice is None:
        df_known["class"] = df_known['Label']
    else:
        raise Exception('Pick a valid class or None')

    y = df_known["class"]
    X = df_known.drop(["Label", "class"], axis=1)    

    return X, y
```

```python
def abbreviate(word):
        '''Replace long names with shorter ones.'''
#         replacements = {'minimum': 'min.', 'maximum': 'max.', 'relative': 'rel.', 'variation': 'var.',
#                         'mean absolute deviation': 'M.A.D.', 'length': 'L.',
#                    'absolute': 'abs.', 'Number': 'Num.', 'Electronegativity': 'El.Neg.',
#                    'deviation': 'dev.', 'CovalentRadius': 'Cov.Rad.', 'distance': 'dist.',
#                        'neighbor': 'N', 'volume': 'vol', 'MagpieData': ''}
        replacements = {'MagpieData ': ''}
    
        for item, abbrv in replacements.items():
            word = word.replace(item, abbrv)
        return word

def abbreviate_features(X):
    """Renames columns in DataFrame to make plots fit better."""
    orig_names = list(X)
    new_names = [abbreviate(word) for word in orig_names]
    column_names = dict(zip(orig_names, new_names))
    X.rename(index=str, columns=column_names, inplace=True)
    return X
```

```python
def remove_redundant_features(X):
    """Drops zero variance and highly correlated features from DataFrame."""
    variances = X.var()
    to_drop = [column for column in variances.index if variances[column] == 0]
    X.drop(columns=to_drop, inplace=True)
    
    # Create correlation matrix
    corr_matrix = X.corr().abs() # Remove text variables before finding correlations

    # Select upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

    # Find index of feature columns with correlation greater than 0.95
    to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

    X.drop(columns=to_drop, inplace=True)
    return X
```

```python
def load_data(choice='MIT', path="../data/IMT_Classification_Dataset_matminer_and_handbuilt_v2.xlsx"):
    """Reads in DataFrame, abbreviates columns, and removes highly correlated features."""
    X, y = read_data(choice, path)
    X = mean_imputation(X)
    X = abbreviate_features(X)
    X = remove_redundant_features(X)
    return X, y
```

# Utility functions

```python
def give_metrics(cv, model, choice):
    """Returns array of precision, recall, and AUC scores for given CV scheme, task, and model.
    
    Params:
        cv: Object, sklearn KFold or StratifiedKFold cross-validator
        model: Object, sklearn or XGBoost model 
        choice: Str, 'MIT', 'Insulator', 'Metal', or None
    Returns:
        results, Tuple of lists of metrics 
    """
    X, y = load_data(choice=choice)
    if len(y.value_counts()) == 2:
        averaging_method = 'binary'
    else:
        averaging_method = 'weighted'
    precisions = []
    recalls = []
    aucs = []
    f1s = []
    
    for train, test in cv.split(X.values, y.values):
        model.fit(X.drop(columns=['Compound','struct_file_path']).values[train], y.values[train])
        # Estimate AUC score if binary classification
        if len(y.value_counts()) == 2:
            y_pred = model.predict_proba(X.drop(columns=['Compound','struct_file_path']).values[test])[:,1]
            aucs.append(roc_auc_score(y.values[test],y_pred))

        # Estimate precision
        y_pred = model.predict(X.drop(columns=['Compound','struct_file_path']).values[test])
        precisions.append(precision_score(y.values[test], y_pred, average=averaging_method))
        
        # Estimate recall
        recalls.append(recall_score(y.values[test], y_pred, average=averaging_method))
        
        # Estimate F1
        f1s.append(f1_score(y.values[test], y_pred, average=averaging_method))
        
    results = (precisions, recalls, f1s, aucs)    
    return results       
```

```python
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a, axis=0), scipy.stats.sem(a, axis=0)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h
```

# Grid search for optimal hyperparameters 

```python
from sklearn.model_selection import GridSearchCV
```

```python
def grid_search():
    """Conducts grid search to find optimal model hyperparameters for all 3 classes."""
    model = XGBClassifier()
    param_grid = {'n_estimators': [10, 20, 40, 80, 100, 150, 200], 
                  'scale_pos_weight': [1, 2, 3],
                 'max_depth': [2, 3, 5],
                 'learning_rate': np.logspace(-3, 2, num=6),
                 'base_score': [0.3, 0.5, 0.7]}
    cv = StratifiedKFold(n_splits=5, shuffle=True)
    params = {}
    for name in ['Metal', 'Insulator', 'MIT']:
        xgb = GridSearchCV(model, param_grid, scoring='f1', cv=cv, n_jobs=5)
        X, y = load_data(choice=name)
        X_train, X_test, y_train, y_test = train_test_split(
        X.drop(columns=['Compound','struct_file_path']), y, test_size=0.2)
        t0 = time.time()
        xgb.fit(X_train, y_train)
        xgb_fit = time.time() - t0
        print("XGB parameters for %s selected in %.3f s"
              % (name, xgb_fit))
        params[name] = xgb.best_params_
    return params
```

# Stratified k-fold cross-validation

```python
averaging_method = 'macro'
precisions = []
recalls = []
aucs = []
f1s = []
for choice in ['Metal', 'Insulator', 'MIT']:
#     model = XGBClassifier(**best_params[choice])
    print('\nFor {} binary classification'.format(choice))
    for test_size in [1.0 / n for n in range(10, 11)]:
        for seed in range(10):
#             model = GradientBoostingClassifier(random_state=seed, n_estimators=100)
            model = XGBClassifier(**best_params[choice])
            cv = StratifiedKFold(n_splits=int(test_size**-1), random_state=seed, shuffle=True)

            prec, rec, f1, auc_score = give_metrics(cv=cv, model=model, choice=choice)
            precisions.extend(prec)
            recalls.extend(rec)
            aucs.extend(auc_score)
            f1s.extend(f1)

    
        print('\nFor {} folds'.format(1 / test_size))
        print('Mean AUC score: {:0.2f} w/std.: {:0.2f}'.format(
            np.mean(aucs), np.std(aucs)))
        print("Mean precision: {:0.2f} w/std.: {:0.2f}".format(
            np.mean(precisions), np.std(precisions)))
        print("Mean recall: {:0.2f} w/std.: {:0.2f}".format(
            np.mean(recalls), np.std(recalls)))
        print("Mean F1: {:0.2f} w/std.: {:0.2f}".format(
            np.mean(f1s), np.std(f1s)))
```

## Precision-recall curves from cross-validation

```python
state = 42
cv = KFold(n_splits=10, random_state=state, shuffle=True)
classes = ['Metal', 'Insulator', 'MIT']
mean_rec = np.linspace(0, 1, 100)
def get_precisions_f1s(n):
    choice = classes[n]
    model = XGBClassifier(**best_params[choice])
    precisions = []
    aucs = []
    f1s = []
    X, y = load_data(choice=choice)
    X = X.drop(columns=['Compound','struct_file_path'])
    X = X.values
    y = y.values
    for train, test in cv.split(X, y):
        probas_ = model.fit(X[train], y[train]).predict_proba(X[test])
        # Compute precision-recall curve
        precision, recall, thresholds = precision_recall_curve(y[test], probas_[:, 1], pos_label=1)
        precision = precision[::-1]
        recall = recall[::-1]
        precisions.append(interp(mean_rec, recall, precision))
        precisions[-1][0] = 1.0
        pr_auc = auc(recall, precision)
        aucs.append(pr_auc)
        f1s.append(f1_score(y[test], model.predict(X[test])))
    return precisions, aucs, f1s

stuff = [get_precisions_f1s(n) for n in range(3)]
```

## Reduce data dimensionality with UMAP before clustering

```python
import umap
```

```python
def generate_embeddings(state):
    """Returns embeddings for each class as dictionary."""
    embeddings = {}
    for name in ['Metal', 'Insulator', 'MIT']:
        rawX, _ = load_data(choice=name)
        X = rawX.drop(columns=['Compound', 'struct_file_path'])
        vals = X.values
        scaler = preprocessing.RobustScaler()
        scaled_vals = scaler.fit_transform(vals)
        X = pd.DataFrame(scaled_vals, columns=X.columns)
        embedding = umap.UMAP(n_neighbors=10,
                              n_components=10, 
                              metric='manhattan',
                             random_state=state,
                             transform_seed=state,
                             min_dist=0.1).fit_transform(X.values)
        embeddings[name] = embedding
    return embeddings
```

```python
def get_clusters_noumap(k):
    data = {}
    for name in ['Metal', 'Insulator', 'MIT']:
        X, y = load_data(choice=name) 
        scaler = preprocessing.StandardScaler()
        X_scaled = scaler.fit_transform(X.drop(columns=['Compound', 'struct_file_path']))
        
        # Perform clustering
        km = KMeans(n_clusters=k, n_init=20)
        km.fit(X_scaled)
#         clusters = km.predict(X_scaled)
        X_scaled['cluster'] = pd.Series(km.predict(X_scaled))
        y_copy = pd.DataFrame(y.values, columns=['Class'])
        y_copy['cluster'] = X_scaled['cluster']
        data[name] = (X_scaled, y_copy)
    return data
```

```python
def get_clusters(k, embeddings):
    """Returns k clusters in embeddings from k means."""
    # Standardize data for k-means clustering
    data = {}
    for name in ['Metal', 'Insulator', 'MIT']:
        _, y = load_data(choice=name) 
        embedding = embeddings[name]
        scaler = preprocessing.StandardScaler()
        scaled_vals = scaler.fit_transform(embedding)
        
        X_scaled = pd.DataFrame(scaled_vals, columns=['Embedding{}'.format(i) for i in range(embedding.shape[1])])
    
        # Perform clustering
        km = KMeans(n_clusters=k, n_init=20)
        km.fit(X_scaled)
        clusters = km.predict(X_scaled)
        X_scaled['cluster'] = km.predict(X_scaled)
        y_copy = pd.DataFrame(y.values, columns=['Class'])
        y_copy['cluster'] = X_scaled['cluster']
        data[name] = (X_scaled, y_copy)
    return data
```

```python
from sklearn.metrics import confusion_matrix
import itertools
X, y = load_data(choice=None)
# X, y = load_data(choice='Metal')
test_size = 0.15
y_tests = []
y_preds = []
model = XGBClassifier(**xgb.best_params_)
for seed in range(10):
    X_train, X_test, y_train, y_test = train_test_split(
        X.drop(columns=['Compound', 'struct_file_path']), y, test_size=test_size, random_state=seed)
    y_tests.extend(y_test)
    model.fit(X_train, y_train)
    class_names = ["Metal", "Insulator", 'MIT']
#     class_names = ["Metal", "Insulator\n+\nMIT"]
    y_pred = model.predict(X_test)
    y_preds.extend(y_pred)

# print(pd.concat([y, y_pred], axis=1))
```

```python
cnfm = confusion_matrix(y_tests, y_preds)
def plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=36)
#     plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)
    plt.tick_params(axis='both', which='major', labelsize=30)
    plt.tick_params(axis='both', which='minor', labelsize=4)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 1.25
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, '{0:.2f}'.format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black",
                 size=36)

    
    plt.ylabel('True label', fontsize=32)
    plt.xlabel('Predicted label', fontsize=32)
    
plt.figure(figsize=(14,12), dpi=400)
plt.gcf().subplots_adjust(bottom=0.15)
plot_confusion_matrix(cnfm, classes=class_names, normalize=True,
                      title='', cmap='Blues')
plt.tight_layout(pad=1.1)
# plt.savefig("../plots/paper_figures/metalvsall_confusion_matrix.pdf", dpi=400)
# plt.savefig("../plots/paper_figures/confusion_matrix.pdf", dpi=400)
# plt.savefig("../plots/confusion_matrix.png", dpi=300)
plt.show()
```
